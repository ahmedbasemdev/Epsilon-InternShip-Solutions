## Business Understanding
**<a href="https://eg.hatla2ee.com/en">Htla2ee.com</a>** is website for selling and buying cars, in this project we will focus on the used car prices in Egyptian market, we try to Analyize the market and get insights about how the used cars get priced and what are the factors that affects the price, also we will make a ML model to predict the car pprice of used car based on its features.
