```python
from selenium import webdriver
import joblib
import numpy as np
import pandas as pd
import csv
from datetime import datetime
import time
import sqlite3
```

```python
driver = webdriver.Chrome()
```

```python
url = 'https://eg.hatla2ee.com/en/car'
driver.get(url)

pages_count = int(driver.find_elements_by_class_name('paginate')[-2].text)
pages_count
```

```python
links = []

for page in range(1, pages_count + 1):
    print(f'Scraping from page: {page}')
    url = f'https://eg.hatla2ee.com/en/car/page/{page}'
    driver.get(url)
    for x in driver.find_elements_by_class_name('newCarListUnit_header'):
        car_url = x.find_element_by_tag_name('a').get_attribute('href')
        links.append(car_url)
        print(car_url)
    print('-'*50)
    
    
links
```

```python
joblib.dump(links, 'links.joblib')
```

```python
links = joblib.load('links.joblib')
links
```

```python
len(links)
```

```python
main_feats = ['Make','Model','CC','Used since','Km','Transmission','City','Color','Body Style','Fuel']
equipments = ['Leather seats','Closing mirrors',
'Intelligent parking system','Sunroof','Fabric brushes','Air Conditioning','Power Steering',
'Remote Keyless','Electric mirrors','Front Power Windows',
'Back Power Windows','Tinted Glass','CD Player','DVD Player','Bluetooth','Multifunction',
'Cassette Radio','AUX','CD Changer','USB Port','Anti - theft System','Side Airbag',
'ABS','EPS','EBD','ESP','Sensors','Rear sensors','Driver Airbag','Passenger Airbag','Front sensors','Alloy wheels','Rear camera','GPS','Fog light','Rear spoiler',
'Cruise control','Power Seats','Central lock','Alarm']

columns = main_feats + equipments + ['Price']
columns
```

```python
error_urls = []
```

```python
import csv

with open('cars_data.csv', mode='w') as csv_file:
    fieldnames = columns
    writer = csv.DictWriter(csv_file, fieldnames=fieldnames)
    writer.writeheader()
    
    for car_url in links[:30]:
        try:
            print(f'Scraping car: {car_url}')
            driver.get(car_url)
            car = {}
            price = driver.find_element_by_class_name('usedUnitCarPrice').text
            car['Price'] = price
            all_blocks = {}
            for block in driver.find_elements_by_class_name('DescDataItem'):
                try:
                    feature = block.find_element_by_class_name('DescDataSubTit').text
                    value = block.find_element_by_class_name('DescDataVal').text
                    all_blocks[feature] = value
                except:
                    continue
            for f in main_feats:
                car[f] = all_blocks.get(f, np.nan)
            for eq in equipments:
                car[eq] = 1 if eq in driver.page_source else 0
            writer.writerow(car)
        except:
            error_urls.append(car_url)
```
