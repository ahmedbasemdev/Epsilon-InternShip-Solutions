**Analyize Numerical columns (Km, Price)**

- As we see Km & Price columns have wrong datatype so lets fix it.

```python
df['Km'].unique()
```

```python
def fix_km(km_str):
    try:
        return int(km_str[:-3].replace(',', ''))
    except:
        return np.nan
    
df['Km'] = df['Km'].apply(fix_km)
```

```python
df['Km'].describe()
```

```python
df['Km'] = df['Km'].apply(lambda km: km if km > 1000 else np.nan)
```

```python
df['Km'].describe()
```

```python
df['Price'].unique()
```

```python
def fix_price(price_str):
    try:
        return int(price_str[:-4].replace(',', ''))
    except:
        return np.nan
    
df['Price'] = df['Price'].apply(fix_price)
```

```python
df['Price'].describe()
```

**min price for a car is 0 EGP so i think that is wrong, lets see the cars below 10K (good price for a very cheap car in egypt)**

```python
df[df['Price'] < 8000]
```

```python
df.drop(df[df['Price'] < 8000].index, inplace=True)
```

```python
df['Price'].describe()
```

**Analyize Categorical columns (Make, Model, CC, Used since, Transmission, City, Color, Body Style, Fuel).**

- As we see Km & Price columns have wrong datatype so lets fix it.

```python
for col in ['Make','Model','CC','Used since','Transmission','City','Color','Body Style','Fuel']:
    print(f'{col}')
    print(f'{df[col].unique()}')
    print('-'*50)
```

```python
def fix_merged_data(v):
    try:
        if ' || ' in v:
            val1, val2 = v.split(' || ')
            return val1 if len(val1) > len(val2) else val2
        else:
            return v
    except:
        return v
    
fix_merged_data('Canghe || Changhe')
```

```python
df['Make'] = df['Make'].apply(fix_merged_data)
df['Model'] = df['Model'].apply(fix_merged_data)
```

**Analyize Missing values.**

```python
df.isna().sum() / df.shape[0] * 100
```

**Lets start handling missing values by:**
- Remove CC & Body Style columns
- Remove the rest rows that has missing values.

```python
df.drop(['CC', 'Body Style'], axis=1, inplace=True)
```

```python
df.isna().sum() / df.shape[0] * 100
```

```python
df
```

```python
from sklearn.impute import SimpleImputer

imputer = SimpleImputer(strategy='median')
df['Km'] = imputer.fit_transform(df[['Km']])
```

```python
df.isna().sum() / df.shape[0] * 100
```

```python
df.dropna(inplace=True)
```
