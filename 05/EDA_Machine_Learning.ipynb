### Machine Learning Part

### We will use cross validation to estimate the best classification model 

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from xgboost import XGBClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_validate, cross_val_predict
```

```python
models = []
models.append(('LogisticRegression', LogisticRegression()))
models.append(('Naive Bayes',GaussianNB()))
models.append(('RandomForest', RandomForestClassifier()))
models.append(('Decision Tree', DecisionTreeClassifier()))
models.append(('KNN', KNeighborsClassifier(n_neighbors = 5)))
models.append(('XGB', XGBClassifier())) 
```

```python
results=[]
names=[]
for name, model in models:
    kfold = KFold(n_splits=10, random_state=7,shuffle=True)
    cv_results = cross_validate(model, x_train, y_train, cv=kfold , return_train_score=True, scoring=('accuracy', 'precision', 'recall'))
    results.append(cv_results)
    names.append(name)
    msg = "%s: %f " % (name, cv_results['train_accuracy'].mean() * 100)
    print("Average_Train_Acc",msg)
    msg = "%s: %f " % (name, cv_results['test_accuracy'].mean()* 100)
    print("Average_Test_Acc",msg)
    score = (2* (cv_results['train_precision'].mean() *  cv_results['train_recall'].mean()) / (cv_results['train_precision'].mean() +  cv_results['train_recall'].mean()) )
    msg = "%s: %f " % (name, score)
    print("Average_Train_F1_Score",msg)
    score = (2* (cv_results['test_precision'].mean() *  cv_results['test_recall'].mean()) / (cv_results['test_precision'].mean() +  cv_results['test_recall'].mean()) )
    msg = "%s: %f " % (name, score)
    print("Average_Test_F1_Score",msg)
    print('\n')
```

#### We can see Descision Trees , Random Forests and XGboost are the best ones but some of them may overfit , so let' s
#### test them on the test set also.

```python
models = []
models.append(('RandomForest', RandomForestClassifier()))
models.append(('Decision Tree', DecisionTreeClassifier()))
models.append(('XGB', XGBClassifier()))
```

```python
for name, model in models:
    print(name)
    model.fit(x_train, y_train)
    
    # Make predictions.
    predictions = model.predict(x_test)

    # Compute the error.
    print(confusion_matrix(predictions, y_test))
    print(accuracy_score(predictions,y_test))
    print('\n')
```

### Now we can see that descision tree isn't overfitting the test data set , so it's very good model , we will use it.

```python
Final_Model = DecisionTreeClassifier()
Final_Model.fit(x_train , y_train)
Predictions = Final_Model.predict(x_test)
Confusion_Matrix =confusion_matrix(Predictions, y_test) 
Accuracy = accuracy_score(Predictions,y_test) * 100
print(Confusion_Matrix)
print(Accuracy)
```
